Replacing blanks with 0 as tenure is 0 and no total charges are recorded

Converted 0 and 1 of senior citizen column to yes/no to make it easier to understand

```python
def conv(value):
    if value==1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
ax=sns.countplot(x="Churn",data=df)
plt.title("Count of Customers by Churn")
ax.bar_label(ax.containers[0])
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index,autopct='%1.1f%%',shadow=True)
plt.axis('equal') 
plt.title("Percentage of Churned Customers",fontsize=10)
```

From the given pie chart we conclude that 26.54% of our customers have churned out 
Now let's explore the reason behind it

```python
plt.figure(figsize=(4,4))
gd=sns.countplot(x="gender",data=df,hue="Churn")
gd.bar_label(gd.containers[0])
gd.bar_label(gd.containers[1])
plt.title("Churn based on Gender",fontsize=10)
```

```python
plt.figure(figsize=(4,4))
gd=sns.countplot(x="SeniorCitizen",data=df)
gd.bar_label(gd.containers[0])
plt.title("Count of Customers by Senior citizen",fontsize=10)
```

```python
total_counts = (
    df.groupby('SeniorCitizen')['Churn']
    .value_counts(normalize=True)  
    .unstack(fill_value=0) * 100  
)
fig, ax = plt.subplots(figsize=(5,5))

total_counts.plot(
    kind='bar',
    stacked=True,
    ax=ax,
    color=['#1f77b4', '#ff7f0e'],  # Custom colors
    width=0.6
)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    
    if height > 0:  
        ax.text(
            x + width / 2,
            y + height / 2,
            f'{height:.1f}%',
            ha='center',
            va='center',
            fontsize=10
        )
plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.legend(title='Churn', loc='upper right')
plt.tight_layout()
```

Comparatively a greater percentage of people in senior citizen category have churned

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure",data=df,bins=50,hue="Churn")
```

People who have used our services for a long time have stayed and people who used our services for 1 or 2 months have churned

```python
plt.figure(figsize=(4,4))
gd=sns.countplot(x="Contract",data=df,hue="Churn")
gd.bar_label(gd.containers[0])
gd.bar_label(gd.containers[1])
plt.title("Count of Customers by Contract",fontsize=10)
```

People who have a contract of month to month have more chances to churn than those with one year or two year contracts

```python
df.columns.values
```

```python


# Step 1: Define the categorical columns
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Step 2: Decide grid layout automatically
n_cols = 3  # number of plots per row
n_rows = (len(columns) + n_cols - 1) // n_cols  # calculate number of rows

# Step 3: Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
axes = axes.flatten()  # flatten 2D array to 1D for easy looping

# Step 4: Generate countplots for each column
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue="Churn")  # default colors, no palette

    # Add count labels to each bar
    for container in axes[i].containers:
        axes[i].bar_label(container, fontsize=8)

    # Set title and labels
    axes[i].set_title(f"Count Plot of {col}", fontsize=12)
    axes[i].set_xlabel("")   # remove x-axis label for clean look
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=15)  # rotate labels for clarity

# Step 5: Remove extra empty plots (if grid > number of columns)
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Step 6: Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

The visualizations show customer distribution across different services like phone, internet, and streaming. Features such as Fiber Optic internet, lack of Online Security, and lack of Tech Support appear strongly linked with higher churn rates. Customers without additional services like Device Protection or Streaming options tend to stay, while those with these services have more mixed churn behavior. This highlights which service types may influence customer retention.
